```python
# AnÃ¡lise de Risco e Estoque de EmergÃªncia - Pandas & NumPy
# Problema: Quais medicamentos priorizar no estoque hospitalar?

import pandas as pd
import numpy as np

# ============================================================================
# 1. IMPORTAÃ‡ÃƒO E CARREGAMENTO
# ============================================================================
df = pd.read_csv('../data/allergies.csv')
print(f"Dataset: {df.shape[0]} registros x {df.shape[1]} colunas\n")
print(df.head(3))
```

```python
# 2. NUMPY ARRAYS E INDEXAÃ‡ÃƒO
# ============================================================================
severity_array = df['SEVERITY1'].dropna().values
print(f"\n{'='*70}")
print("NUMPY ARRAYS E INDEXAÃ‡ÃƒO")
print('='*70)
print(f"Array NumPy: {len(severity_array)} elementos")
print(f"Primeiros 5: {severity_array[:5]}")
print(f"Ãšltimos 5: {severity_array[-5:]}")
print(f"Elementos 10-15: {severity_array[10:15]}")
```

```python
# 3. MÃSCARAS BOOLEANAS
# ============================================================================
print(f"\n{'='*70}")
print("MÃSCARAS BOOLEANAS")
print('='*70)
mask_severe = severity_array == 'SEVERE'
mask_moderate = severity_array == 'MODERATE'
mask_critical = mask_severe | mask_moderate

print(f"Casos SEVEROS: {mask_severe.sum()}")
print(f"Casos MODERATE: {mask_moderate.sum()}")
print(f"Casos CRÃTICOS (SEVERE OU MODERATE): {mask_critical.sum()}")
```

```python
# 4. FANCY INDEXING
# ============================================================================
print(f"\n{'='*70}")
print("FANCY INDEXING")
print('='*70)
random_indices = np.array([0, 10, 20, 30, 90])
print(f"Ãndices selecionados: {random_indices}")
print(f"Valores correspondentes: {severity_array[random_indices]}")
```

```python
# ============================================================================
# 5. UFUNCS - OPERAÃ‡Ã•ES VETORIZADAS
# ============================================================================
print(f"\n{'='*70}")
print("UFUNCS - OPERAÃ‡Ã•ES VETORIZADAS")
print('='*70)
# SimulaÃ§Ã£o de dados de diferentes medicamentos
patient_counts = np.array([45, 32, 78, 12, 56, 23, 91, 34])
severe_counts = np.array([5, 2, 15, 1, 8, 3, 20, 4])

# CÃ¡lculo vetorizado de risk scores
risk_scores = patient_counts + (severe_counts * 10)
print(f"Risk Scores: {risk_scores}")
print(f"MÃ©dia: {np.mean(risk_scores):.2f}")
print(f"Mediana: {np.median(risk_scores):.2f}")
print(f"Desvio padrÃ£o: {np.std(risk_scores):.2f}")
print(f"Score mÃ¡ximo: {np.max(risk_scores)}")
print(f"Score mÃ­nimo: {np.min(risk_scores)}")
```

```python
# ============================================================================
# 6. AGREGAÃ‡Ã•ES NUMPY
# ============================================================================
print(f"\n{'='*70}")
print("AGREGAÃ‡Ã•ES NUMPY")
print('='*70)
categories = df['CATEGORY'].dropna().values 
unique_cats, counts = np.unique(categories, return_counts=True)

print("Contagem por categoria:")
for cat, count in zip(unique_cats, counts):
    print(f"  {cat}: {count}")
```

```python
 # ============================================================================
# 7. ÃLGEBRA LINEAR - SCORES PONDERADOS
# ============================================================================
print(f"\n{'='*70}")
print("ÃLGEBRA LINEAR - SCORES PONDERADOS")
print('='*70)
# Matrix: [frequÃªncia, severidade, peso_categoria]
features = np.array([
    [45, 5, 3],
    [32, 2, 2],
    [78, 15, 3],
    [12, 29, 17]
])

weights = np.array([1, 3, 7])
weighted_scores = np.dot(features, weights)

print(f"Scores ponderados: {weighted_scores}")

# NormalizaÃ§Ã£o 0-100
normalized = (weighted_scores - weighted_scores.min()) / (weighted_scores.max() - weighted_scores.min()) * 100
print(f"Scores normalizados (0-100): {normalized.round(2)}")
```

```python
# 8. PANDAS SERIES
# ============================================================================
print(f"\n{'='*70}")
print("PANDAS SERIES")
print('='*70)
severity_series = pd.Series(
    ['SEVERE', 'MODERATE', 'MILD', 'SEVERE', 'MODERATE'],
    index=['P1', 'P2', 'P3', 'P4', 'P5'],
    name='Severidade'
)

print("Series:")
print(severity_series)
print("\nValue counts:")
print(severity_series.value_counts())
```

```python
# ============================================================================
# 9. DATAFRAME - TIPOS E INFO
# ============================================================================
print(f"\n{'='*70}")
print("DATAFRAME - TIPOS E INFO")
print('='*70)
print("\nTipos de dados:")
print(df.dtypes)

print(f"\nShape: {df.shape}")
print(f"Colunas: {list(df.columns)}")
```

```python
# ============================================================================
# 10. CONVERSÃƒO DE TIPOS
# ============================================================================
print(f"\n{'='*70}")
print("CONVERSÃƒO DE TIPOS")
print('='*70)
df['START'] = pd.to_datetime(df['START'], errors='coerce')
print(f"Tipo de START apÃ³s conversÃ£o: {df['START'].dtype}")

# CORREÃ‡ÃƒO: Verificar se as colunas existem antes
if 'TYPE' in df.columns:
    print(f"\nCategorias em TYPE: {df['TYPE'].unique()}")
if 'CATEGORY' in df.columns:
    print(f"Categorias em CATEGORY: {df['CATEGORY'].unique()}")
```

```python
# ============================================================================
# 11. ÃNDICES
# ============================================================================
print(f"\n{'='*70}")
print("MANIPULAÃ‡ÃƒO DE ÃNDICES")
print('='*70)
print(f"Ãndice original: {type(df.index)}")

df_indexed = df.set_index('PATIENT')
print(f"Ãndice por PATIENT: {df_indexed.index.name}")

df_reset = df_indexed.reset_index()
print(f"Ãndice resetado: {type(df_reset.index)}")
```

```python
# ============================================================================
# 12. SELEÃ‡ÃƒO - loc, iloc, at, iat
# ============================================================================
print(f"\n{'='*70}")
print("SELEÃ‡ÃƒO - loc, iloc, at, iat")
print('='*70)
print("\nloc - por rÃ³tulos:")
print(df.loc[0:2, ['DESCRIPTION', 'CATEGORY', 'SEVERITY1']])

print("\niloc - por posiÃ§Ã£o:")
print(df.iloc[0:3, [6, 8, 10]])

print(f"\nat - valor escalar: {df.at[0, 'DESCRIPTION']}")
print(f"iat - valor escalar: {df.iat[0, 6]}")
```

```python
# ============================================================================
# 13. FILTROS BOOLEANOS
# ============================================================================
print(f"\n{'='*70}")
print("FILTROS BOOLEANOS")
print('='*70)
severe_df = df[df['SEVERITY1'] == 'SEVERE']
print(f"Alergias SEVERAS: {len(severe_df)}")
print(severe_df[['DESCRIPTION', 'CATEGORY']].head(3))

severe_meds = df[(df['SEVERITY1'] == 'SEVERE') & (df['CATEGORY'] == 'medication')]
print(f"\nMedicamentos SEVEROS: {len(severe_meds)}")

critical_cats = df[df['CATEGORY'].isin(['medication', 'food'])]
print(f"Medicamentos ou alimentos: {len(critical_cats)}")
```

```python
# ============================================================================
# 14. SETTINGWITHCOPYWARNING - BOAS PRÃTICAS
# ============================================================================
print(f"\n{'='*70}")
print("BOAS PRÃTICAS - EVITAR SETTINGWITHCOPYWARNING")
print('='*70)
# Correto: usar .copy()
df_food = df[df['CATEGORY'] == 'food'].copy()
df_food['RISK_LEVEL'] = 'HIGH'
print(f"DataFrame copiado: {len(df_food)} registros com RISK_LEVEL")

# Correto: usar .loc no original
df.loc[df['CATEGORY'] == 'food', 'IS_FOOD'] = True
print("Coluna IS_FOOD adicionada ao DataFrame original")
```

```python
# ============================================================================
# 15. GROUPBY
# ============================================================================
print(f"\n{'='*70}")
print("GROUPBY - AGREGAÃ‡Ã•ES POR CATEGORIA")
print('='*70)
category_counts = df.groupby('CATEGORY').size()
print("\nContagem por categoria:")
print(category_counts)

category_agg = df.groupby('CATEGORY').agg({
    'PATIENT': ['count', 'nunique']
})
category_agg.columns = ['Total_Casos', 'Pacientes_Unicos']
print("\nAgregaÃ§Ãµes mÃºltiplas:")
print(category_agg)

severity_by_cat = df.groupby(['CATEGORY', 'SEVERITY1']).size().unstack(fill_value=0)
print("\nSeveridade por categoria:")
print(severity_by_cat)
```

```python
# ============================================================================
# 16. PIVOT TABLES
# ============================================================================
print(f"\n{'='*70}")
print("PIVOT TABLES")
print('='*70)
pivot = pd.pivot_table(
    df,
    values='PATIENT',
    index='CATEGORY',
    columns='SEVERITY1',
    aggfunc='count',
    fill_value=0,
    margins=True
)
print("Pivot com totais:")
print(pivot)
```

```python
# ============================================================================
# 17. CONCAT
# ============================================================================
print(f"\n{'='*70}")
print("CONCAT - CONCATENAÃ‡ÃƒO DE DATAFRAMES")
print('='*70)
df_med = df[df['CATEGORY'] == 'medication'].head(2)
df_food = df[df['CATEGORY'] == 'food'].head(2)

df_concat = pd.concat([df_med, df_food], axis=0, ignore_index=True)
print(f"Concat vertical: {len(df_concat)} registros")
print(df_concat[['DESCRIPTION', 'CATEGORY']])
```

```python
# ============================================================================
# 18. MERGE
# ============================================================================
print(f"\n{'='*70}")
print("MERGE - ENRIQUECIMENTO DE DADOS")
print('='*70)
category_info = pd.DataFrame({
    'CATEGORY': ['medication', 'food', 'environment'],
    'PRIORITY': ['ALTA', 'ALTA', 'MÃ‰DIA'],
    'KIT_REQUIRED': [True, True, False]
})

df_enriched = df.merge(category_info, on='CATEGORY', how='left')
print("Merge - dados enriquecidos:")
print(df_enriched[['DESCRIPTION', 'CATEGORY', 'PRIORITY']].head(5))
```

```python
# ============================================================================
# 19. DATAS E TEMPOS
# ============================================================================
print(f"\n{'='*70}")
print("ANÃLISE TEMPORAL")
print('='*70)
df['YEAR'] = df['START'].dt.year
df['MONTH'] = df['START'].dt.month
df['QUARTER'] = df['START'].dt.quarter

print("Componentes de data:")
print(df[['START', 'YEAR', 'MONTH', 'QUARTER']].head(3))

by_year = df.groupby('YEAR').size().sort_index()
print("\nAlergias por ano:")
print(by_year.tail(10))
```

```python
# ============================================================================
# 20. ROLLING WINDOWS
# ============================================================================
print(f"\n{'='*70}")
print("ROLLING WINDOWS - TENDÃŠNCIAS TEMPORAIS")
print('='*70)
monthly = df.sort_values('START').groupby(df['START'].dt.to_period('M')).size()
ts_df = pd.DataFrame({
    'Month': monthly.index.to_timestamp(),
    'Count': monthly.values
})

ts_df['Rolling_3M'] = ts_df['Count'].rolling(3, min_periods=1).mean()
ts_df['Cumsum'] = ts_df['Count'].cumsum()

print("AnÃ¡lise temporal com rolling:")
print(ts_df.tail(10))
```

```python
# ============================================================================
# 21. RESOLUÃ‡ÃƒO DO PROBLEMA DE NEGÃ“CIO
# ============================================================================
print("\n" + "="*70)
print("RESOLUÃ‡ÃƒO: PRIORIZAÃ‡ÃƒO DE ESTOQUE DE EMERGÃŠNCIA")
print("="*70)

# PERGUNTA 1: Quantas alergias foram registradas no sistema?
print(f"\n{'â”'*70}")
print("PERGUNTA 1: Quantas alergias foram registradas no sistema?")
print('â”'*70)
df_allergies = df[df['TYPE'] == 'allergy'].copy()
total_alergias = len(df_allergies)
print(f"âœ“ RESPOSTA:")
print(f"  â€¢ {total_alergias:,} alergias registradas no sistema")
print(f"  â€¢ Representa {total_alergias/len(df)*100:.1f}% do dataset completo")

# PERGUNTA 2: Qual a distribuiÃ§Ã£o de severidade das alergias?
print(f"\n{'â”'*70}")
print("PERGUNTA 2: Qual a distribuiÃ§Ã£o de severidade das alergias?")
print('â”'*70)

# Calcular severidade mÃ¡xima usando NumPy para melhor performance
def max_severity(row):
    sevs = [row['SEVERITY1'], row['SEVERITY2']]
    sevs = [s for s in sevs if pd.notna(s)]
    if 'SEVERE' in sevs: 
        return 'SEVERE'
    elif 'MODERATE' in sevs: 
        return 'MODERATE'
    elif 'MILD' in sevs: 
        return 'MILD'
    return 'UNKNOWN'

df_allergies['MAX_SEVERITY'] = df_allergies.apply(max_severity, axis=1)

severity_dist = df_allergies['MAX_SEVERITY'].value_counts()
print("âœ“ RESPOSTA:")
for sev, count in severity_dist.items():
    pct = count/len(df_allergies)*100
    print(f"  â€¢ {sev}: {count:,} casos ({pct:.1f}%)")

# PERGUNTA 3: Quais categorias de alergia sÃ£o mais crÃ­ticas?
print(f"\n{'â”'*70}")
print("PERGUNTA 3: Quais categorias de alergia sÃ£o mais crÃ­ticas?")
print('â”'*70)

risk_data = []
for cat in ['medication', 'food', 'environment']:
    subset = df_allergies[df_allergies['CATEGORY'] == cat]
    total = len(subset)
    severe = int(np.sum(subset['MAX_SEVERITY'] == 'SEVERE'))
    moderate = int(np.sum(subset['MAX_SEVERITY'] == 'MODERATE'))
    mild = int(np.sum(subset['MAX_SEVERITY'] == 'MILD'))
    # Score: cada caso conta 1, casos severos +10, moderados +3
    score = total + severe * 10 + moderate * 3
    risk_data.append([cat, total, severe, moderate, mild, score])

df_risk = pd.DataFrame(
    risk_data, 
    columns=['Categoria', 'Total', 'Severos', 'Moderados', 'Leves', 'Score']
)
df_risk = df_risk.sort_values('Score', ascending=False)

print("âœ“ RESPOSTA:")
print(df_risk.to_string(index=False))
print(f"\n  ğŸ¥‡ Categoria mais crÃ­tica: {df_risk.iloc[0]['Categoria']} (Score: {int(df_risk.iloc[0]['Score'])})")

# PERGUNTA 4: Quais sÃ£o as 10 alergias mais perigosas?
print(f"\n{'â”'*70}")
print("PERGUNTA 4: Quais sÃ£o as 10 alergias mais perigosas?")
print('â”'*70)

allergy_summary = df_allergies.groupby(['DESCRIPTION', 'CATEGORY']).agg({
    'PATIENT': 'count',
    'MAX_SEVERITY': lambda x: int(np.sum(x == 'SEVERE'))
}).rename(columns={'PATIENT': 'Total', 'MAX_SEVERITY': 'Severos'})

allergy_summary['Score'] = allergy_summary['Total'] + allergy_summary['Severos'] * 10
top10 = allergy_summary.sort_values('Score', ascending=False).head(10).reset_index()

print("âœ“ RESPOSTA:")
for idx, row in top10.iterrows():
    print(f"  {idx+1:2d}. {row['DESCRIPTION'][:50]:<50} | {row['CATEGORY']:<12} | "
          f"Total: {row['Total']:3d} | Severos: {row['Severos']:3d} | Score: {row['Score']:4d}")

# PERGUNTA 5: Quais medicamentos devem ser priorizados no estoque?
print(f"\n{'â”'*70}")
print("PERGUNTA 5: Quais medicamentos devem ser priorizados no estoque?")
print('â”'*70)

meds = df_allergies[df_allergies['CATEGORY'] == 'medication'].copy()
med_summary = meds.groupby('DESCRIPTION').agg({
    'PATIENT': 'count',
    'MAX_SEVERITY': [
        ('Severos', lambda x: int(np.sum(x == 'SEVERE'))),
        ('Moderados', lambda x: int(np.sum(x == 'MODERATE'))),
        ('Leves', lambda x: int(np.sum(x == 'MILD')))
    ]
})

# Flatten multi-level columns
med_summary.columns = ['Total', 'Severos', 'Moderados', 'Leves']
med_summary['Score'] = med_summary['Total'] + med_summary['Severos'] * 10 + med_summary['Moderados'] * 3
top_meds = med_summary.sort_values('Score', ascending=False).head(10)

print("âœ“ RESPOSTA:")
for idx, (med, row) in enumerate(top_meds.iterrows(), 1):
    print(f"  {idx:2d}. {med[:45]:<45} | Score: {row['Score']:4d}")
    print(f"      Total: {row['Total']:3d} | Severos: {row['Severos']:3d} | "
          f"Moderados: {row['Moderados']:3d} | Leves: {row['Leves']:3d}")

# PERGUNTA 6: Quantos pacientes Ãºnicos sÃ£o afetados?
print(f"\n{'â”'*70}")
print("PERGUNTA 6: Quantos pacientes Ãºnicos sÃ£o afetados?")
print('â”'*70)

pacientes_unicos = df_allergies['PATIENT'].nunique()
media_alergias_por_paciente = len(df_allergies) / pacientes_unicos

print("âœ“ RESPOSTA:")
print(f"  â€¢ Pacientes Ãºnicos com alergias: {pacientes_unicos:,}")
print(f"  â€¢ MÃ©dia de alergias por paciente: {media_alergias_por_paciente:.2f}")

# Pacientes com mÃºltiplas alergias severas
pacientes_severos = df_allergies[df_allergies['MAX_SEVERITY'] == 'SEVERE'].groupby('PATIENT').size()
multi_severos = (pacientes_severos > 1).sum()
print(f"  â€¢ Pacientes com mÃºltiplas alergias SEVERAS: {multi_severos:,}")

# PERGUNTA 7: Existe tendÃªncia temporal no registro de alergias?
print(f"\n{'â”'*70}")
print("PERGUNTA 7: Existe tendÃªncia temporal no registro de alergias?")
print('â”'*70)

# Verificar se as colunas de data foram criadas
if 'YEAR' not in df_allergies.columns:
    df_allergies['YEAR'] = df_allergies['START'].dt.year
    df_allergies['MONTH'] = df_allergies['START'].dt.month
    df_allergies['QUARTER'] = df_allergies['START'].dt.quarter

# AnÃ¡lise anual
yearly = df_allergies.groupby('YEAR').size().sort_index()
crescimento = ((yearly.iloc[-1] - yearly.iloc[0]) / yearly.iloc[0] * 100) if len(yearly) > 1 else 0

print("âœ“ RESPOSTA:")
print(f"  â€¢ PerÃ­odo analisado: {int(yearly.index.min())} a {int(yearly.index.max())}")
print(f"  â€¢ Registros no primeiro ano: {yearly.iloc[0]:,}")
print(f"  â€¢ Registros no Ãºltimo ano: {yearly.iloc[-1]:,}")
print(f"  â€¢ VariaÃ§Ã£o: {crescimento:+.1f}%")

print("\n  EvoluÃ§Ã£o anual (Ãºltimos 5 anos):")
for year, count in yearly.tail(5).items():
    year_int = int(year) if not pd.isna(year) else 0
    bar = 'â–ˆ' * int(count / yearly.max() * 50)
    print(f"    {year_int}: {bar} {count:,}")

# PERGUNTA 8: Quais alergias alimentares sÃ£o mais comuns?
print(f"\n{'â”'*70}")
print("PERGUNTA 8: Quais alergias alimentares sÃ£o mais comuns?")
print('â”'*70)

food_allergies = df_allergies[df_allergies['CATEGORY'] == 'food']
top_food = food_allergies.groupby('DESCRIPTION').size().sort_values(ascending=False).head(5)

print("âœ“ RESPOSTA:")
for idx, (food, count) in enumerate(top_food.items(), 1):
    pct = count / len(food_allergies) * 100
    print(f"  {idx}. {food[:50]:<50} | {count:4d} casos ({pct:.1f}%)")

# PERGUNTA 9: Qual a taxa de criticidade por categoria?
print(f"\n{'â”'*70}")
print("PERGUNTA 9: Qual a taxa de criticidade (SEVERE+MODERATE) por categoria?")
print('â”'*70)

print("âœ“ RESPOSTA:")
for cat in ['medication', 'food', 'environment']:
    subset = df_allergies[df_allergies['CATEGORY'] == cat]
    if len(subset) > 0:
        critical = int(np.sum((subset['MAX_SEVERITY'] == 'SEVERE') | 
                             (subset['MAX_SEVERITY'] == 'MODERATE')))
        taxa = critical / len(subset) * 100
        nivel = "ğŸ”´ ALTA" if taxa > 20 else "ğŸŸ¡ MÃ‰DIA" if taxa > 10 else "ğŸŸ¢ BAIXA"
        print(f"  {cat:<15} | {critical:4d}/{len(subset):4d} crÃ­ticos | "
              f"Taxa: {taxa:5.1f}% | {nivel}")
```

```python
# ## 22. RECOMENDAÃ‡Ã•ES FINAIS

print("\n" + "="*70)
print("RECOMENDAÃ‡Ã•ES BASEADAS EM DADOS")
print("="*70)

total_severe = np.sum(df_allergies['MAX_SEVERITY'] == 'SEVERE')
total_moderate = np.sum(df_allergies['MAX_SEVERITY'] == 'MODERATE')

print(f"\nESTATÃSTICAS:")
print(f"  â€¢ Total alergias: {len(df_allergies):,}")
print(f"  â€¢ Casos SEVEROS: {total_severe:,} ({total_severe/len(df_allergies)*100:.1f}%)")
print(f"  â€¢ Casos MODERADOS: {total_moderate:,} ({total_moderate/len(df_allergies)*100:.1f}%)")

print(f"\nPRIORIDADES DE ESTOQUE:")
for _, row in df_risk.iterrows():
    priority = "ğŸ”´ ALTA" if row['Score'] > 500 else "ğŸŸ¡ MÃ‰DIA" if row['Score'] > 200 else "ğŸŸ¢ BAIXA"
    print(f"  {priority} - {row['Categoria']}: Score {row['Score']}")

print(f"\nMEDICAMENTOS ESSENCIAIS:")
print(f"  1. EPINEFRINA - {total_severe} casos severos")
print(f"  2. ANTI-HISTAMÃNICOS - {total_moderate} casos moderados")
print(f"  3. CORTICOSTEROIDES - {total_severe + total_moderate} casos crÃ­ticos")

print(f"\nTOP 5 ALERTAS CRÃTICOS:")
for idx, (med, row) in enumerate(top_meds.head(5).iterrows(), 1):
    print(f"  {idx}. {med}: {row['Severos']} severos, {row['Moderados']} moderados")

print("\n" + "="*70)
print("âœ… ANÃLISE CONCLUÃDA")
print("="*70)

print("""
CONCEITOS DEMONSTRADOS:
âœ“ NumPy: Arrays, Slicing, MÃ¡scaras, Fancy Indexing, Ufuncs, Ãlgebra Linear
âœ“ Pandas: Series, DataFrame, Tipos, Ãndices, loc/iloc/at/iat
âœ“ ManipulaÃ§Ã£o: Filtros, GroupBy, Pivot, Concat, Merge
âœ“ Temporal: Datas, Rolling Windows
âœ“ ResoluÃ§Ã£o: Problema de negÃ³cio com dados reais
""")
```
